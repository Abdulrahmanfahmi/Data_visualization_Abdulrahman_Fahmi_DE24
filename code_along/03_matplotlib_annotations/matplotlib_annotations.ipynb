```python
duckdb.query("""
    WITH world AS (
        SELECT
            'USA' AS country,
            SUM("Total foreign born") AS total_foreigners,
            SUM(Sweden) AS swedes,
            SUM(Denmark) AS danes,
            SUM(Finland) AS finns,
            SUM(Norway) AS norwegians,
            SUM(Australia) AS australians,
            SUM(Austria) AS austrians,
            SUM(Belgium) AS belgians,
            SUM(Turkey) AS turks,
            SUM(Spain) AS spanish
        FROM df_world
    )
    SELECT * FROM world
    ORDER BY 
        world 
    DESC
""").df()

df_world.head()
```

```python
countries = ["Sweden", "Denmark", "Finland", "Norway", "Australia", "Austria", "Belgium", "Turkey", "Spain"]


df[countries] = df[countries].apply(pd.to_numeric, errors="coerce")


filtered_df = df[df[countries].gt(0).any(axis=1)]

cities = filtered_df[["City", "State"]].drop_duplicates().sort_values(by="City")
print(cities)
```
