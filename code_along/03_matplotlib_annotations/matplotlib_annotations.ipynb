```python
import matplotlib.pyplot as plt


ax = top_3_world.plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title('Nordisk befolkning i olika städer')
plt.ylabel('Antal personer')
plt.xlabel('Stad, Stat')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
def add_annotation(ax, city, stat, values):
    
    # denna raden hittar stapelns position
    idx = list(top_3_world.index).index((city,stat))
    
    # denna raden beräknar total höjd för staplarna
    
    total_height = sum(v for v in values if pd.notan(v))
    
    # annoteringstext
    text = f"{city}\noTotalt: {int(total_height)}"
    
    ax.annotate(
        text=text,
        xy=(idx, total_height),  # Toppen av staplarna
        xytext=(idx+0.3, total_height*1.1),  # Position för text
        arrowprops=dict(facecolor='black', shrink=0.05, width=2, headwidth=8),
        bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.8),
        ha='center'
    )
    
```
