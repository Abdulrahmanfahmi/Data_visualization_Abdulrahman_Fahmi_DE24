```python
import duckdb

df_quantity_brand = duckdb.query("""
    SELECT 
        make AS brand,
        SUM (quantity) AS Quantity
    FROM df
    GROUP BY
        make
    ORDER BY
        quantity
    DESC             
""").df()

df_quantity_brand.head()
```

## Bar chart

# Matplotlib bar

```python
import matplotlib.pyplot as plt

df_quantity_brand_top10 = df_quantity_brand.iloc[:8]

fig, ax = plt.subplots()
```

```python
import matplotlib.pyplot as plt

df_quantity_brand_top10 = df_quantity_brand.iloc[:8]

fig, ax = plt.subplots()

ax.bar(
    x = df_quantity_brand_top10 ["brand"],
    height = df_quantity_brand_top10 ["Quantity"]
)
fig.tight_layout()

fig.savefig("figures/bars_top_8_cars.png")
```

```python
import matplotlib.pyplot as plt

df_quantity_brand_top10 = df_quantity_brand.iloc[:8]

fig, ax = plt.subplots(figsize = (12,7))

ax.bar(
    x = df_quantity_brand_top10 ["brand"],
    height = df_quantity_brand_top10 ["Quantity"],
    width=0.7,
)
```

```python
import matplotlib.pyplot as plt

df_quantity_brand_top10 = df_quantity_brand.iloc[:8]

fig, ax = plt.subplots(figsize = (12,7))

ax.bar(
    x = df_quantity_brand_top10 ["brand"],
    height = df_quantity_brand_top10 ["Quantity"],
    width=0.7,
)

ax.set(
    title = "Number of cars from top 10 most sold brands in Norway 2007-2017",
    xlabel = "Car brand",
    ylabel = "Number of cars"
)

fig.tight_layout()

fig.savefig("figures/bars_top_10_cars.png")
```
