```python
df_salary = (
    duckdb.query(
        """
    SELECT 
        job_category, 
        CAST(AVG(salary_in_usd) AS INT) AS avg_salary_usd
    FROM 
        df
    GROUP BY 
        job_category
    ORDER BY avg_salary_usd DESC             
"""
    )
    .df()
    .set_index("job_category")
)

df_salary
```

```python
df_salary.plot(kind="barh")
```

```python
colors = [GRAY_1]*len(df_salary)
colors[3] = BLUE_1

df_salary["avg_salary_usd"].plot(kind="barh",color=colors)
```
